## Introdução

Nesta semana, iremos mergulhar nos fundamentos de Ciência de Dados (DC) e Inteligência Artificial (IA). Além disso, exploraremos a utilização da biblioteca Pandas para ciência de dados e as etapas necessárias para elaborar um projeto de ciência de dados bem-sucedido.

## Data Science

### Introdução

Começaremos discutindo os conceitos de dados, informações e conhecimento. Atualmente, os dados têm um valor significativo na sociedade e no mundo dos negócios. Compreender o contexto atual dos dados é essencial para identificar e aproveitar novas oportunidades no mercado de trabalho.

### Dado

Dado é o menor elemento que define algo. Um conjunto de dados processados gera uma informação. Informações catalogadas formam o conhecimento e um grupo de conhecimentos é a inteligência (sabedoria). 

É importante avaliar o nível de maturidade da empresa. Ela ainda está na era do "processamento de dados" ou já avançou para a "gestão do conhecimento"?

### Informação

Entrada, Processamento e Saída são as principais etapas do processamento de dados.

### Conhecimento

Com o aumento dos dados, surge uma nova categoria chamada "Big Data", e com ela a profissão de mineração de conhecimento (Data Mining). Esta área de mineração de dados tem se associado cada vez mais à ciência de dados.

### Ciência de Dados é Novo?

Não. A ciência de dados é uma área antiga que recebeu um "upgrade" com a ciência da computação. A constante atualização com ferramentas, linguagens de programação e bibliotecas poderosas, aliadas a ótimos frameworks de mercado, modernizou a área. É neste contexto que você está agora: aprendendo uma área tradicional, mas modernizada e valorizada com o avanço do Big Data e da Inteligência Artificial.

## Etapas para DC e IA

Iniciaremos enfrentando um problema como exercício. Entender o problema é o primeiro passo. Posteriormente, vamos aplicar a biblioteca Pandas para carregar e tratar os dados. Após a preparação e validação dos dados, seguiremos técnicas de um projeto de inteligência artificial, como escolher o tipo de aprendizagem, o melhor algoritmo, separar dados de teste e de treino, treinar e realizar otimizações finais.

### Definindo o Problema

Você deve definir o problema e coletar os dados relevantes. Após a coleta, os dados devem ser organizados e estruturados em um dataset (no nosso caso, um CSV). O objetivo final é utilizar esses dados em uma inteligência artificial, seja por aprendizado supervisionado, não supervisionado ou por reforço.

## Carreiras

### Analista de Dados

Porta de entrada para a ciência de dados. Trabalha com melhoria de processos, relatórios para gestores, correção de problemas relacionados aos dados e ao código, entre outras tarefas.

### Cientista de Dados

Utiliza diversas áreas como ciência da computação, matemática e estatística. Com um olhar científico, utiliza ferramentas e linguagens de programação específicas para tratar dados, preparando-os para que o engenheiro de machine learning possa trabalhar com o dataset com o mínimo de falhas possível. Geralmente, programa em Python, R e Julia.

### Engenheiro de Dados

Organiza e estrutura dados provenientes de diversas fontes, uma tarefa complexa que requer uma área específica para trabalhar com dados organizados: a engenharia de dados.

### Engenheiro de Machine Learning

Após os dados estarem tratados e prontos para treino, o engenheiro de machine learning escolhe algoritmos e utiliza técnicas de otimização para desenvolver modelos preditivos eficazes.
