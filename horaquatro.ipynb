# Qual o valor médio de aluguel por tipo de imóvel?

Tpo e valor são as duas colunas mais importantes.

Calcular  a média

```python
#seleciona a coluna valor

dados['Valor'].mean()
```

### Agrupar - usando o método groupby você pode agrupar. Coloque o nome da coluna que deseja que seja agrupada. Esse método trabalha junto com as funções de agregação.

Temos colunas númericas e não númericas. 

```python
dados.groupby('Tipo').mean(numeric_only=True)
```

Ele retornou para a gente um dataframe, o indice o tipo de imoveis. E trouxe a quantidade média de quartas, vagas, suites e o que realmente importa é o valor médio de aluguel.

# Vamos agora pegar só o valor.

Antes o método mean() colocamos a seleção para a coluna Valor.

```python
dados.groupby('Tipo')['Valor'].mean()
```

Agora sim conseguimos mostrar apenas a média do valor dos tipos de imóveis. Ele não precisou do numeric_
Retornou uma Series
Onde os índices são os tipos de imóveis e valores as médias desses tipos.

Vamos tranformar em um dataframe usando [[ ]]
Posteriormente, podemos ordenar, no caso colocamos a coluna valor.

```python
dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

# Podemos plotar um gráfico.

barh - barras horizontais
tamanho do gráfico
e cor.

```python
df_preco_tipo = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')

df_preco_tipo.plot(kind="barh",figsize=(14,10),color='purple')
```

Os imóveis estão oscilando muito, e  o time de ML quer um base que seja ideal para aprendizado do modelo. Imóveis residenciais e comercial acabam não trazendo valores compatíveis entre ambos. Uma técnica importante é retirar os imoaveis comerciais dos residênciais e assim ficará melhor para o time de ML.

# Método Query da Biblioteca Pandas

Vamos percorrer apenas os imóveis comerciais.

```python
dados.query('@imoveis_comerciais in Tipo')
```

Vamos passar para o métoodo uma expressão. Arroba imoveis_comerciais, estou dizendo para o query selecionar apenas os componentes dos imóveis comerciais. O arroba, é para especificar uma variável em uma expressão.

# O que queremos não é realmente o oposto disso?

```python
dados.query('@imoveis_comerciais not in Tipo')
```

# Vamos agora guardar em uma variável

```python
df = dados.query('@imoveis_comerciais not in Tipo')
df.head()
```

```python
df.Tipo.unique()
```

Veja que agora tenho bem menos tipos de imóveis.

# Qual o percentual de cada tipo de imóvel na nossa base de dados?

```python
df.Tipo.unique()
```

```python
df.Tipo.value_counts()
```

Eu preciso ver essas quantidades em porcentagens

```python
df.Tipo.value_counts(normalize=True)
```

Vamos converter essa Serie em um Dataframe

```python
df.Tipo.value_counts(normalize=True).to_frame()
```

Vamos ordenar? Ordenando por Tipo

```python
df.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo')
```

```python
df.Tipo.value_counts(normalize=True).to_frame().sort_values('proportion')
```

# Vamos plotar?

```python
df_percentual_tipo=df.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo')

df_percentual_tipo.plot(kind="bar",figsize=(14,10),color='green',xlabel='Tipos',ylabel='Percentual')
```

Os imóveis do tipo apartamento esta pegando 80 por cento de apartamento.
